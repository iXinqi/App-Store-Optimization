```python
import time
import requests
import sys ,os,math
import json
import datetime
import requests
from collections import Counter

import io
import urllib
import http.cookiejar

appname = '网易彩票'
appidset = ['1322932843','1357552825','1171321498','1095996950','1344702233']        #提取上一步中所有的appid，形成竞品列表
```

```python
time1 = time.time()
```

```python
timeout = 4
def get_comp_prod(appname):
    url ='http://rest.appbk.com/app/get_all_app_search_results?n=' + appname + '&start=0&limit=100&cc=cn' 
    appid = []
    try:
        resp_ = requests.get(url,timeout=timeout,verify = False)
        resp_2 = resp_.content
        resp_3 = resp_2.decode('utf-8')
        
        resp_list = json.loads(resp_3)
        resp_list = resp_list['results']
        for item in resp_list:
            appid.append(item['app_id']) 
        return appid    
    except Exception as e:
        print(e)
```

```python
appidset = get_comp_prod(appname)
appidset = appidset[0:100] 
if len(appidset):
    for item in appidset:
        if item not in appidset:
            appidset.append(item)
```

```python
timeout = 4
def wordcover_appbk(appid):
    today = datetime.date.today().strftime('%Y-%m-%d')
    yesterday = (datetime.date.today()+datetime.timedelta(days=-1)).strftime('%Y-%m-%d')
    url ='http://rest.appbk.com/word/get_app_word_two_date_compare?email=&app_id=' + appid + '&token=&start=0&limit=12000&simple=1&cur_date=' + today + '&pre_date=' + yesterday
    try:
        resp_ = requests.get(url,timeout=timeout,verify = False)
        resp_2 = resp_.content
        resp_3 = resp_2.decode('utf-8')

        resp_list = json.loads(resp_3)
    
        hotness = []
        key = []
        resp_list = resp_list['results']
        print(resp_list)
        for item in resp_list:
            if int(item[3]) >= 4605:   # and int(item[1]) <= 3
                key.append(item[0]+'@@@'+str(appid))
                hotness.append(int(item[3]))
        return dict(zip(key, hotness))
    except Exception as e:
        print(e)
    return -1
```

```python
wordcover_dict = {}
for appid in appidset:
    wordcover_dict = dict(wordcover_dict, **wordcover_appbk(appid))
```

```python
name_rank = []
for item in list(wordcover_dict.items()):
    word = item[0].split('@@@')[0]
    item = (word, item[1])
    name_rank.append(item)
```

```python
countlist = Counter(name_rank).most_common(200)
wordslist = []
hotlist = []
for i in countlist:
    wordslist.append(i[0][0])
    hotlist.append(i[0][1])
wordhot_dict = dict(zip(wordslist, hotlist))
wordhot_dict = sorted(wordhot_dict.items(),key = lambda x:x[1],reverse = True)
```

```python
words = ''
for item in wordhot_dict:
    words += item[0]
words
```

```python
######   组词（推荐方案）

from jieba import posseg
from functools import reduce
import jieba

def remove_dual(corpus):
    func = lambda x,y:x if y in x else x + [y]
    keyword_list = reduce(func, [[], ] + corpus)
    return keyword_list

def count_len(wordlist):
    listlen = 0
    for i in wordlist:
        listlen += len(i)
    return listlen

seg_list = posseg.cut(words)

corpus=[]
for word,flag in seg_list:
    if flag != 'x':    
        corpus.append(word)

corpus = remove_dual(corpus)

keyword = ''
strlen = 0
for line in corpus:    
    keyword += line 
    strlen += len(line)
    if strlen>15:
        keyword += ', '
        strlen = 0
keyword = keyword[0:100]
keyword
```
